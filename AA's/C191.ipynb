```python
import  numpy as np
import pandas as pd 
import matplotlib .pyplot as plt

dataframe = pd.read_csv("space_Corrected.csv")
df=dataframe.dropna()

df.head(10)

```

```python
#Teacher-Activity-1- displaying the information of the dataframe columns 
df.info()

#Teacher-Activity-2- removing the unwanted columns from the dataset. So that from this dataset student can start working
del df['Unnamed: 0.1']
del df['Unnamed: 0']

df
```

```python
#Student-Activity-1
#Find total number of mission by each company, and plot a bar graph on it

group_by_name = df.groupby('Company Name')['Status Mission'].count().reset_index(name='Total Mission')
print(group_by_name)
fig = plt.subplots(figsize=(16,8))
plt.title('Total Missions (Since 1957)', fontsize=20)
plt.xlabel('Company Name', fontsize=16)
plt.ylabel('Mission Counts', fontsize=16)
plt.xticks(rotation='vertical')

label = group_by_name['Company Name']
value = group_by_name['Total Mission']
plt.bar(label, value,width=0.4, color=('red','blue','green','pink','yellow')) #bar-grap
```

```python
#Student-Activity-2
#Find out the percentage of rocket Success, Failure, Partial Failure, and Prelaunch Failure. And plot a pie chart

group_by_status = df.groupby('Status Mission')['Status Mission'].count().reset_index(name='Number')
print(group_by_status)
value = group_by_status['Number']
label = group_by_status['Status Mission']
plt.pie(value,labels=label, autopct='%0.1f%%', radius=2)
plt.show()
```

```python
#Student-Activity-3
# Find the number of Active Rockets as per the company and plot a bar grap of it
search_StatusActive= df.loc[df['Status Rocket'] == 'StatusActive']

group_by_status = search_StatusActive.groupby('Company Name')['Status Rocket'].count().reset_index(name='Status')
print(group_by_status)
fig = plt.subplots(figsize=(16,8))
plt.xlabel("Company Name")
plt.xticks(rotation='vertical')
plt.ylabel("Number Of Active Rockets")

label = group_by_status['Company Name'] 
value = group_by_status['Status']

plt.bar(label, value,width=0.4, color=('red','blue','green','pink','yellow')) #bar-grap
```
